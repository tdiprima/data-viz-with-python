## 1. What've we got?

### List the csv files in a directory

```python
import os
import fnmatch

# directory = '/path/to/directory'
directory = '.'

for file in os.listdir(directory):
    if fnmatch.fnmatch(file, '*.csv'):  # Check if the file has a .csv extension
        print(os.path.join(directory, file))  # Print the file path
```

## 2. Load CSV file into Pandas DataFrame

```python
# Load CSV file into a Pandas DataFrame and get the file name
file_name = 'gas_prices.csv'
df_gas_prices = pd.read_csv(file_name)

# Print the file name
print("File name:", file_name)
```

## Scatter plot

A scatter plot is best used to show the relationship between two continuous variables and identify any patterns or clusters in the data.

Scatter plots are particularly useful for visualizing the correlation between two variables, which can be positive, negative, or zero. They can also be used to identify clusters or patterns in the data.

```python
import os
import sys

try:
    # Scatter plot (2) columns
    df.plot(x='Height', y='Weight', kind='scatter')
    plt.title('Scatter plot of Height vs Weight')
    plt.xlabel('Height')
    plt.ylabel('Weight')
    plt.show()
except Exception as e:
    exc_type, exc_obj, exc_tb = sys.exc_info()
    print("Line:", exc_tb.tb_lineno)
    print("Error:", exc_type, exc_obj)
```
